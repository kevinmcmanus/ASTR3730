The largest spike is the dominant frequency in the dataset, so you want to find *what index this occurs at* (ie what place it is on the x-axis). Recall that the x-axis here is a frequency space, which is spaced basically (technically depends on whether function is odd or even) like $f = \frac{[0, 1, 2, ...]}{\Delta t N}$ where $\Delta t$ is the difference in time between each datapoint and N is the total number of data points. 

The largest spike is the dominant frequency in the dataset, so you want to find *what index this occurs at* (ie what place it is on the x-axis). Recall that the x-axis here is a frequency space, which is spaced like $f = \frac{[0, 1, 2, ...]}{\Delta t N}$ where $\Delta t$ is the difference in time between each datapoint and N is the total number of data points. The data stops being useful at index $\approx \frac{N}{2}$ as the FFT is symmetric (the exact spot it becomes not useful depends on whether N is odd or even). 

If that's confusing and you don't want to think about it, numpy has a function for that too! You can call something like: 

```python
freq = np.fft.fftfreq(len(sunspots),1/12) #first argument is number of data points, second is dt
```

And this will generate an array of frequencies that correspond to your data — so if you found the peak at index 42 you could find out what physical frequency that corresponds to by doing something like 

```python
domFreq = freq[42]
```

````python
import astropy.units as u

spfft = np.fft.fft(sunspots.N_sunspots)
spcoefs = np.abs(spfft)
coefs = spcoefs[1:len(spcoefs)//2]

t = np.arange(len(sunspots)) # in months

maxpower_i = coefs.argmax()
freqs = np.fft.fftfreq(len(sunspots), (1/12)*u.year)[1:len(spcoefs)//2]
maxpower_i = coefs.argmax()
maxpower_freq = 1/freqs[maxpower_i]
print(f'Dominant Frequency is: {maxpower_freq:.2f}')
````

````python

xlim=(freqs[0].value,freqs[100].value)
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot()

ax.axvline(freqs[maxpower_i].value, color='red', label=f'Dominant Frequency: {maxpower_freq:.2f}')
ax.plot(freqs, coefs, label='FFT Coefficents')
ax.legend()
ax.set_xlim(xlim)
````

## Fit Periodic Function


$y=Asin^2(\omega t + \phi_0)$ 

From [Wikipedia](https://en.wikipedia.org/wiki/Solar_cycle):
>Until 2009, it was thought that 28 cycles had spanned the 309 years between 1699 and 2008, giving an average length of 11.04 years, but research then showed that the longest of these (1784–1799) may actually have been two cycles.[14][15] If so then the average length would be only around 10.7 years.

````python
from scipy.optimize import minimize
````

````python
# functions for scipy.optimize.minimize():

def ssr(params, y, t):
    #calculates sum of square difference btwn y and yhat
    yhat = calculate_yhat(params, t)
    e2 = ((y - yhat)**2).sum()
    return e2

def calculate_yhat(params, t):
    A = params[0]; omega = params[1]; phi = params[2]
    s2 = np.sin((np.pi/(12*omega))*t+phi)**2
    return A*s2



````

````python
xlim=(len(sunspots)-600, len(sunspots)) #last fifty years
#xlim=None
cl=11.25; amp = 150; phi = 0.8*np.pi
init_guess = np.array([amp, cl, phi])
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot()
ax.plot(t, sunspots.N_sunspots, label='Actual')
ax.axhline(sunspots.N_sunspots.mean(), color='red', label='mean')
ax.plot(t, calculate_yhat(init_guess, t), color='black',label='Guessed')
ax.legend()

ax.set_title('Sunspot Activity')
ax.set_xlabel('Months since $t_0$')
ax.set_ylabel('Sunspots per Month')

ax.set_xlim(xlim)
print(f'Parameters: Cycle Length: {cl} years, Max Amp: {amp}, phi: {np.degrees(phi):.2f} degrees')
````

````python

params = minimize(ssr, init_guess, (sunspots.N_sunspots,t),  options={'maxiter':500} )
params
````

````python
fig = plt.figure(figsize=(16,5))
ax = fig.add_subplot()
fitted_params = params['x']
cl=fitted_params[1]; amp = fitted_params[0]; phi = fitted_params[2]
ax.plot(t, sunspots.N_sunspots, label='Actual')
ax.axhline(sunspots.N_sunspots.mean(), color='red', label='mean')
ax.plot(t, calculate_yhat(fitted_params, t), color='black',label='Fitted')
ax.legend()

ax.set_title('Sunspot Activity')
ax.set_xlabel('Months since $t_0$')
ax.set_ylabel('Sunspots per Month')

#ax.set_xlim(xlim)
print(f'Parameters: Cycle Length: {cl:.2f} years, Max Amp: {amp:.2f}, phi: {np.degrees(phi):.2f} degrees')
````
